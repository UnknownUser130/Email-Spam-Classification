# ScamAssasin Dataset

## Easy Ham

```python
import os
import email
from email import policy

# Path to the extracted directory
ham_dir = 'easy_ham'

# List all files in the directory
files = os.listdir(ham_dir)

# Load the emails
emails = []
for file_name in files:
    with open(os.path.join(ham_dir, file_name), 'r', encoding='latin1') as file:
        msg = email.message_from_file(file, policy=policy.default)
        # Extract features
        subject = msg['subject']
        from_ = msg['from']
        to = msg['to']
        date = msg['date']
        body = msg.get_body(preferencelist=('plain', 'html')).get_content()
        # Append to the emails list
        emails.append({'subject': subject, 'from': from_, 'to': to, 'date': date, 'body': body, 'label': '0'})

# Convert to DataFrame if needed
import pandas as pd

ham_df = pd.DataFrame(emails)

# Optionally, save to a CSV for further use
ham_df.to_csv('easy_ham.csv', index=False)
```

## Spam

```python
import os
import email
from email import policy

# Path to the extracted directory
ham_dir = 'spam'

# List all files in the directory
files = os.listdir(ham_dir)

# Load the emails
emails = []
for file_name in files:
    file_path = os.path.join(ham_dir, file_name)
    with open(file_path, 'r', encoding='latin1') as file:
        try:
            # Parse the email message
            msg = email.message_from_file(file, policy=policy.default)

            # Extract features
            subject = msg['subject']
            from_ = msg['from']
            to = msg['to']
            date = msg['date']

            # Attempt to extract the body content
            try:
                body = msg.get_body(preferencelist=('plain', 'html')).get_content()
            except Exception as e:
                print(f"Error extracting body for {file_name}: {e}")
                body = ""

            # Append to the emails list
            emails.append({'subject': subject, 'from': from_, 'to': to, 'date': date, 'body': body, 'label': '1'})
        except Exception as e:
            print(f"Failed to parse {file_name}: {e}")

# Convert to DataFrame if needed
import pandas as pd

spam_df = pd.DataFrame(emails)

# Optionally, save to a CSV for further use
spam_df.to_csv('spam.csv', index=False)
```
